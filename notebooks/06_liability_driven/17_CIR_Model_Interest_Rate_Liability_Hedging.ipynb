## CIR Model — Interest Rate Scenarios

```python
import ipywidgets as widgets
from IPython.display import display
import matplotlib.pyplot as plt
```

```python
def show_cir(r_0=0.03, a=0.03, b=0.03, sigma=0.05, n_scenarios=5):
    """
    Plot CIR interest rate scenarios (rates only).
    Uses erk.cir() from kit.py which returns (rates, prices).
    """
    fig, ax = plt.subplots(figsize=(12, 6))
    
    rates, _ = erk.cir(r_0=r_0, a=a, b=b, sigma=sigma, n_scenarios=n_scenarios)
    rates.plot(ax=ax, legend=False, lw=0.8, alpha=0.7)
    
    ax.set_title("CIR Interest Rate Scenarios", fontsize=14, fontweight="bold")
    ax.set_xlabel("Month")
    ax.set_ylabel("Annualized Rate")
    ax.axhline(y=b, ls="--", color="red", lw=1.5, label=f"Long-run mean b={b:.2%}")
    ax.legend(loc="upper right")
    ax.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    plt.close(fig)
```

```python
# Interactive widget — rates only
slider_r0    = widgets.FloatSlider(value=0.03, min=0, max=0.15, step=0.01, description="r₀")
slider_a     = widgets.FloatSlider(value=0.03, min=0, max=1, step=0.1,  description="a (speed)")
slider_b     = widgets.FloatSlider(value=0.03, min=0, max=0.15, step=0.01, description="b (mean)")
slider_sigma = widgets.FloatSlider(value=0.05, min=0, max=0.1, step=0.01,  description="σ")
slider_n     = widgets.IntSlider(value=5,   min=1, max=100,             description="Scenarios")

ui = widgets.VBox([slider_r0, slider_a, slider_b, slider_sigma, slider_n])
out = widgets.interactive_output(show_cir, {
    "r_0": slider_r0, "a": slider_a, "b": slider_b, 
    "sigma": slider_sigma, "n_scenarios": slider_n
})
display(ui, out)
```

## CIR Model — Rates & Zero-Coupon Bond Prices

```python
def show_cir_prices_rate(r_0=0.03, a=0.03, b=0.03, sigma=0.05, n_scenarios=5):
    """
    Interactive dual-panel visualization:
      Top    — CIR interest rate paths
      Bottom — Zero-coupon bond prices (analytical CIR formula)
    Uses erk.cir() which returns (rates_df, prices_df).
    """
    rates_df, prices_df = erk.cir(r_0=r_0, a=a, b=b, sigma=sigma, n_scenarios=n_scenarios)
    
    fig, (ax_rate, ax_price) = plt.subplots(2, 1, figsize=(12, 9), sharex=True)
    
    # --- Top: Interest Rates ---
    rates_df.plot(ax=ax_rate, legend=False, lw=0.8, alpha=0.7)
    ax_rate.axhline(y=b, ls="--", color="red", lw=1.5, label=f"Long-run mean b={b:.2%}")
    ax_rate.set_title("CIR Interest Rate Paths", fontsize=13, fontweight="bold")
    ax_rate.set_ylabel("Annualized Rate")
    ax_rate.legend(loc="upper right")
    ax_rate.grid(True, alpha=0.3)
    
    # --- Bottom: Zero-Coupon Bond Prices ---
    prices_df.plot(ax=ax_price, legend=False, lw=0.8, alpha=0.7)
    ax_price.set_title("Zero-Coupon Bond Prices (CIR Analytical)", fontsize=13, fontweight="bold")
    ax_price.set_xlabel("Month")
    ax_price.set_ylabel("Price")
    ax_price.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    plt.close(fig)
```

```python
# Interactive widget — rates & prices
slider_r0_2    = widgets.FloatSlider(value=0.03, min=0, max=0.15, step=0.01, description="r₀")
slider_a_2     = widgets.FloatSlider(value=0.03, min=0, max=1, step=0.1,  description="a (speed)")
slider_b_2     = widgets.FloatSlider(value=0.03, min=0, max=0.15, step=0.01, description="b (mean)")
slider_sigma_2 = widgets.FloatSlider(value=0.05, min=0, max=0.1, step=0.01,  description="σ")
slider_n_2     = widgets.IntSlider(value=5,   min=1, max=100,             description="Scenarios")

ui2 = widgets.VBox([slider_r0_2, slider_a_2, slider_b_2, slider_sigma_2, slider_n_2])
out2 = widgets.interactive_output(show_cir_prices_rate, {
    "r_0": slider_r0_2, "a": slider_a_2, "b": slider_b_2, 
    "sigma": slider_sigma_2, "n_scenarios": slider_n_2
})
display(ui2, out2)
```
