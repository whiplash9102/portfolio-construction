```python
bond_cf_1= erk.bond_cash_flows(15, 1000, coupon_rate=.05, coupon_per_year=2)
bond_cf_2 = erk.bond_cash_flows(5, 1000, coupon_rate=.06, coupon_per_year=4)
flows_B3 = erk.bond_cash_flows(
    maturity=10,
    principal=1000,
    coupon_rate=0.0,
    coupon_per_year=1
)
flows_B3
```

```python
bond_duration_1 = erk.macaulay_duration(bond_cf_1, .05/2)/2
bond_duration_2 = erk.macaulay_duration(bond_cf_2, .05/4)/4
duration_B3 = erk.macaulay_duration(flows_B3, 0.05 / 1) / 1
duration_B3
```
